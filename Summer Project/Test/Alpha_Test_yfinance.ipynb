```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Função para fazer scraping de tickers do S&P 500 da Wikipedia
def scrape_sp500_tickers():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table', {'id': 'constituents'})
    df = pd.read_html(str(table))[0]
    return df['Symbol'].tolist()

# Função para obter tickers de uma API financeira (exemplo: Alpha Vantage)
def get_tickers_from_api(api_key):
    url = f'https://www.alphavantage.co/query?function=LISTING_STATUS&apikey={api_key}&datatype=csv'
    response = requests.get(url)

    # Verificar se a resposta é válida
    if response.status_code != 200:
        print(f"Erro ao acessar a API: {response.status_code}")
        return []

    # Verificar se a resposta está em CSV
    if 'text/csv' in response.headers['Content-Type']:
        lines = response.text.split('\n')
        tickers = [line.split(',')[0] for line in lines[1:] if line]  # O primeiro campo é o símbolo do ticker
        return tickers
    else:
        print("A resposta da API não está no formato CSV.")
        return []

# Substitua 'YOUR_ALPHA_VANTAGE_API_KEY' pela sua chave de API real
api_key = 'YOUR_ALPHA_VANTAGE_API_KEY'
tickers_wikipedia = scrape_sp500_tickers()
tickers_api = get_tickers_from_api(api_key)

# Combinar todas as listas de tickers
all_tickers = set(tickers_wikipedia + tickers_api)

# Mostrar os primeiros 10 tickers
print(list(all_tickers)[:10])
```

```python
# Função para baixar dados históricos de uma ação
def download_data(ticker, period='1d', interval='1m'):
    stock_data = yf.download(ticker, period=period, interval=interval)
    return stock_data
```

```python
# Função para preparar os dados
def prepare_data(data):
    data_selected = data[['Open', 'Close', 'Volume']]
    data_selected['Average'] = data_selected[['Open', 'Close']].mean(axis=1)
    data_final = data_selected[['Volume', 'Average']]
    return data_final
```

```python
# Função para criar as matrizes e os labels
def create_matrices(data_final):
    num_rows = data_final.shape[0]
    num_matrices = num_rows // 5  # Número de matrizes completas que podemos formar

    matrices = []
    matrices_y = []

    for i in range(num_matrices):
        start_idx = i * 5
        end_idx = start_idx + 5
        matrix = data_final.iloc[start_idx:end_idx].copy()

        # Calcular a soma da coluna de volume
        volume_sum = matrix['Volume'].sum()

        # Tornar a coluna do volume relativa
        matrix['Volume'] = matrix['Volume'] / volume_sum

        # Converter para valores numpy e adicionar à lista de matrizes
        matrices.append(matrix.values)

        # Criar labels
        if i > 0:  # Ignorar a primeira matriz para alinhamento correto
            first_value = matrices[i][0, 1]
            last_value = matrices[i][4, 1]
            delta = last_value - first_value
            delta_bi = 1 if delta > 0 else 0
            matrices_y.append(delta_bi)

    # Ordenar as linhas de cada matriz por ordem decrescente da coluna Average (segunda coluna)
    matrices_x = []
    for matrix in matrices:
        sorted_indices = np.argsort(matrix[:, 1])[::-1]
        sorted_matrix = matrix[sorted_indices]
        matrices_x.append(sorted_matrix)

    # Remover a última matriz de matrices_x para alinhamento com matrices_y
    if matrices_x:
        matrices_x.pop()

    # Converter listas para arrays numpy
    matrices_x = np.array(matrices_x)
    matrices_y = np.array(matrices_y)

    return matrices_x, matrices_y
```
