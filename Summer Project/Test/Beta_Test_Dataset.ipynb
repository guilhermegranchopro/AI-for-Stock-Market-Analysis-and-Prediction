```python
import pandas as pd

# Caminho do arquivo Excel (.xlsx) na sua Google Drive
file_path = '/content/drive/MyDrive/AI Financial Analisys/Summer Project/Test/NYSE_and_NYSE_MKT_Trading_Units_Daily_File.xlsx'

# Leitura do arquivo Excel usando Pandas
df = pd.read_excel(file_path)

# Acessar e imprimir os valores da coluna B
coluna_symbol = df['Symbol']  # Seleciona a coluna B
print("Valores da Coluna Symbol:")
print(coluna_symbol)
```

```python
import yfinance as yf

# Função para baixar dados históricos de uma ação
def download_data(ticker, period='max', interval='1m'):
    try:
        stock_data = yf.download(ticker, period=period, interval=interval)
        return stock_data
    except Exception as e:
        print(f"Erro ao baixar dados para {ticker}: {e}")
        return None
```

```python
num_companys = 0
tickers = []
stock_data_list = {}

for i in range(len(coluna_symbol)):
  ticker = coluna_symbol[i]
  stock_data = download_data(ticker, period='max', interval='1m')
  if stock_data is not None:
    stock_data_list[ticker] = stock_data
    tickers.append(ticker)
    num_companys += 1
```

```python
print("Número de empresas: ", num_companys)
print("Tickers:", tickers)
```

```python
import numpy as np

# Function to prepare the data
def prepare_data(data):
    # Selecionar colunas relevantes e fazer uma cópia para evitar advertências
    data_selected = data[['Open', 'Close', 'Volume']].copy()

    # Calcular a média entre 'Open' e 'Close' usando .loc para evitar o SettingWithCopyWarning
    data_selected.loc[:, 'Average'] = data_selected[['Open', 'Close']].mean(axis=1)

    # Selecionar as colunas finais
    data_final = data_selected[['Volume', 'Average']]

    return data_final


def create_matrices(data_final):
    num_rows = data_final.shape[0]
    num_matrices = num_rows // 10  # Número de matrizes completas que podem ser formadas

    if num_matrices == 0:
        raise ValueError("Insufficient data to form a single matrix of 10 rows.")

    matrices = []
    matrices_y = []

    for i in range(num_matrices):
        start_idx = i * 10
        end_idx = start_idx + 10
        matrix = data_final.iloc[start_idx:end_idx].copy()

        # Calcular a soma da coluna Volume
        volume_sum = matrix['Volume'].sum()

        # Normalizar a coluna Volume
        matrix['Volume'] = matrix['Volume'] / volume_sum

        # Calcular a soma da coluna Price (Average)
        price_sum = matrix['Average'].sum()

        # Normalizar a coluna Price
        matrix['Average'] = matrix['Average'] / price_sum

        # Converter para valores numpy e adicionar à lista de matrizes
        matrices.append(matrix.values)

        # Criar rótulos
        if i > 0:  # Ignorar a primeira matriz para alinhamento adequado
            first_value = matrices[i][0, 1]
            last_value = matrices[i][9, 1]  # Comparar a primeira e a última linha
            delta = last_value - first_value
            delta_bi = 1 if delta > 0 else 0
            matrices_y.append(delta_bi)

    # Ordenar as linhas de cada matriz em ordem decrescente pela segunda coluna (Average)
    matrices_x = []
    for matrix in matrices:
        sorted_indices = np.argsort(matrix[:, 1])[::-1]
        sorted_matrix = matrix[sorted_indices]
        matrices_x.append(sorted_matrix)

    # Remover a última matriz de matrices_x para alinhamento com matrices_y
    if matrices_x:
        matrices_x.pop()

    # Converter listas para arrays numpy
    matrices_x = np.array(matrices_x)
    matrices_y = np.array(matrices_y)

    return matrices_x, matrices_y
```

```python
all_matrices_x = []
all_matrices_y = []

# Create matrices and labels for each stock and concatenate them
for ticker in tickers:
    stock_data = stock_data_list[ticker]
    prepared_data = prepare_data(stock_data)
    try:
        matrices_x, matrices_y = create_matrices(prepared_data)

        # Verificar e imprimir as dimensões das matrizes antes de adicionar
        print(f"Shape of matrices_x for {ticker} before adding: {matrices_x.shape}")
        print(f"Shape of matrices_y for {ticker} before adding: {matrices_y.shape}")

        # Filtrar as matrizes que têm a forma (x, 10, 2)
        if matrices_x.ndim == 3 and matrices_x.shape[1:] == (10, 2):
            all_matrices_x.append(matrices_x)
            all_matrices_y.append(matrices_y)
        else:
            print(f"Skipping {ticker} due to incorrect shape of matrices_x: {matrices_x.shape}")
    except ValueError as e:
        print(f"Skipping {ticker}: {e}")
```

```python
# Verificar as dimensões das matrizes antes de concatená-las
for i, matrix in enumerate(all_matrices_x):
    print(f"Shape of matrix {i}: {matrix.shape}")
```

```python
# Concatenate all matrices and labels
try:
    if all_matrices_x:
        all_matrices_x = np.vstack(all_matrices_x)
        all_matrices_y = np.concatenate(all_matrices_y)
except ValueError as e:
    print(f"Error concatenating matrices: {e}")
```

Rede neuronal para um dataset sem importância temporal.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Função para construir a rede neural
def build_model(input_shape):
    model = Sequential([
        Flatten(input_shape=input_shape),
        Dense(64, activation='relu'),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Construir a rede neural
model = build_model(input_shape=(10, 2))

# Treinar o modelo
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val))
```

```python
# Avaliar o modelo nos dados de teste
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Plotar a precisão e a perda ao longo das épocas
import matplotlib.pyplot as plt

# Precisão
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Perda
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

Rede neuronal para um dataset com importância temporal.



```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# Função para construir a rede neural com LSTM
def build_model(input_shape):
    model = Sequential([
        LSTM(64, input_shape=input_shape, return_sequences=True),
        LSTM(64, return_sequences=False),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Construir a rede neural
model = build_model(input_shape=(10, 2))

# Treinar o modelo
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val))
```

```python
# Avaliar o modelo nos dados de teste
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy}")
```
