```python
import requests
from datetime import datetime, timedelta

# Sua chave de API da Alpha Vantage
API_KEY = 'ITHZKPVGZ2QKUXW0'

# Função para obter dados intradiários de uma ação específica
def get_intraday_data(symbol, minutes_ago):
    now = datetime.now()
    ten_minutes_ago = now - timedelta(minutes=minutes_ago)
    ten_minutes_ago_str = ten_minutes_ago.strftime('%Y-%m-%d %H:%M:%S')

    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={symbol}&interval=1min&apikey={API_KEY}'
    response = requests.get(url)
    data = response.json()['Time Series (1min)']

    first_5_minutes = []
    last_5_minutes = []
    count = 0

    for time in sorted(data.keys(), reverse=True):
        if time <= ten_minutes_ago_str:
            count += 1
            if count <= 5:
                values = data[time]
                high = float(values['2. high'])
                low = float(values['3. low'])
                open_price = float(values['1. open'])
                close = float(values['4. close'])
                volume = int(values['5. volume'])
                first_5_minutes.append([high, low, open_price, close, volume, (open_price + close) / 2])
            else:
                values = data[time]
                open_price = float(values['1. open'])
                close = float(values['4. close'])
                last_5_minutes.append(open_price)
                last_5_minutes.append(close)
        else:
            break

    return first_5_minutes, last_5_minutes

# Exemplo de uso
if __name__ == "__main__":
    symbol = 'AAPL'  # Símbolo da ação que você deseja obter dados

    # Obtendo os dados dos primeiros 5 minutos e os últimos 5 minutos restantes do intervalo de 10 minutos atrás
    first_5_minutes, last_5_minutes = get_intraday_data(symbol, 10)

    # Imprimindo a matriz dos primeiros 5 minutos
    print("Primeiros 5 minutos:")
    print("High\tLow\tOpen\tClose\tVolume\tVolume(Relative)\tAvg(Open, Close)")
    for row in first_5_minutes:
        print("\t".join(map(str, row)))

    # Imprimindo os valores de Open e Close dos últimos 5 minutos
    print("\nÚltimos 5 minutos (Open, Close):", last_5_minutes)
```

```python
import requests

def get_stock_data(symbol, api_key):
    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={api_key}"
    response = requests.get(url)
    data = response.json()
    return data

def extract_open_close(data):
    time_series = data.get('Time Series (Daily)', {})
    latest_date = list(time_series.keys())[0]
    open_price = time_series[latest_date]['1. open']
    close_price = time_series[latest_date]['4. close']
    return open_price, close_price

# Substitua 'SUA_CHAVE_API' pela sua chave de API Alpha Vantage
api_key = 'SUA_CHAVE_API'
stock_symbol = 'AAPL'

data = get_stock_data(stock_symbol, api_key)
open_price, close_price = extract_open_close(data)
print(f"Preço de abertura da Apple: {open_price}")
print(f"Preço de fechamento da Apple: {close_price}")
```
