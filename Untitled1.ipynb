```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate some sample data (replace this with your actual data)
np.random.seed(0)
num_candles = 1000
volume = np.random.normal(100, 20, num_candles)
price_open = np.random.normal(50, 10, num_candles)
price_close = np.random.normal(50, 10, num_candles)
price_high = np.random.normal(60, 10, num_candles)
price_low = np.random.normal(40, 10, num_candles)

# Create a DataFrame to hold the data
data = pd.DataFrame({
    'Volume': volume,
    'Open': price_open,
    'Close': price_close,
    'High': price_high,
    'Low': price_low
})

# Calculate some statistics for each candle
data['VolumeMean'] = data['Volume'].rolling(window=20).mean()  # Example: 20-period moving average of volume
data['PriceRange'] = data['High'] - data['Low']

# Define a simple trading strategy
def trading_strategy(row):
    if row['Volume'] > row['VolumeMean'] and row['Close'] < row['Open']:
        return 'Buy'
    elif row['Close'] > row['Open']:
        return 'Sell'
    else:
        return 'Hold'

# Apply the trading strategy to each row
data['Signal'] = data.apply(trading_strategy, axis=1)

# Plot the price and volume data along with the trading signals
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Close Price')
plt.plot(data['Volume'] * 0.1, label='Volume (scaled)', alpha=0.5)
plt.scatter(data.index, data[data['Signal'] == 'Buy']['Close'], marker='^', color='g', label='Buy Signal')
plt.scatter(data.index, data[data['Signal'] == 'Sell']['Close'], marker='v', color='r', label='Sell Signal')
plt.title('Trading Signals')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Fetch historical data for a specific stock (e.g., Apple - AAPL)
ticker_symbol = 'AAPL'
start_date = '2022-01-01'
end_date = '2022-12-31'

data = yf.download(ticker_symbol, start=start_date, end=end_date)

# Calculate rolling mean and standard deviation
window = 5  # Adjust the window size as needed
data['RollingMean'] = data['Close'].rolling(window=window).mean()
data['RollingStd'] = data['Close'].rolling(window=window).std()

# Generate buy and sell signals based on the rolling mean and standard deviation
data['Signal'] = ''
data.loc[data['Close'] < data['RollingMean'] - data['RollingStd'], 'Signal'] = 'Buy'
data.loc[data['Close'] > data['RollingMean'] + data['RollingStd'], 'Signal'] = 'Sell'

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:green'
ax1.set_xlabel('Date')
ax1.set_ylabel('Close Price', color=color)
ax1.plot(data.index, data['Close'], color=color, label='Close Price')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Volume (scaled)', color=color)
ax2.plot(data.index, data['Volume'] * 0.1, color=color, label='Volume (scaled)', alpha=0.5)
ax2.tick_params(axis='y', labelcolor=color)

# Scatter plot for buy and sell signals
ax1.scatter(data[data['Signal'] == 'Sell'].index, data[data['Signal'] == 'Sell']['Close'], marker='v', color='r', label='Sell Signal')
ax1.scatter(data[data['Signal'] == 'Buy'].index, data[data['Signal'] == 'Buy']['Close'], marker='^', color='g', label='Buy Signal')

fig.tight_layout()
plt.title('Trading Signals for {}'.format(ticker_symbol))
plt.legend()
plt.show()

# Calculate balance
balance = 0
stock_count = 0
buy_price = 0

for i in range(len(data)):
    if data['Signal'][i] == 'Buy' and balance == 0:
        balance -= data['Close'][i]  # Deduct cash
        stock_count += 1
        buy_price = data['Close'][i]
    elif data['Signal'][i] == 'Sell' and stock_count > 0:
        balance += stock_count * data['Close'][i]  # Add cash
        balance -= 0.01 * stock_count * data['Close'][i]  # Deduct transaction fee (e.g., 1%)
        stock_count = 0

# If holding any stocks at the end, sell them to get cash
if stock_count > 0:
    balance += stock_count * data['Close'].iloc[-1]
    balance -= 0.01 * stock_count * data['Close'].iloc[-1]

print("Total balance after trading:", balance)
```
