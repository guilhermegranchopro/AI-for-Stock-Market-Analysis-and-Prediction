```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Baixar os dados da Apple (ticker AAPL) para os últimos 7 dias com intervalo de 5 minutos
data = yf.download('AAPL', interval='5m', period='7d')

# Calcular o Volume Profile Delta
data['Delta'] = np.where(data['Close'] > data['Open'], data['Volume'], -data['Volume'])

# Calcular os preços de fechamento únicos
unique_close_prices = data['Close'].unique()

# Calcular o volume delta para cada preço de fechamento
volume_profile_delta = []
for price in unique_close_prices:
    volume_profile_delta.append(data[data['Close'] == price]['Delta'].sum())

# Plotar o Volume Profile Delta Histogram
plt.bar(unique_close_prices, volume_profile_delta, width=0.2)
plt.xlabel('Preço de Fechamento')
plt.ylabel('Volume Profile Delta')
plt.title('Volume Profile Delta Histogram - AAPL')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fetch historical data of S&P 500 using yfinance
sp500 = yf.Ticker("^GSPC")

# Set the start and end dates for the data
start_date = "2024-02-28"
end_date = "2024-03-06"

# Fetch historical data within the specified time frame
data = sp500.history(start=start_date, end=end_date)

# Calculate the daily price change
data['PriceChange'] = data['Close'] - data['Open']

# Classify days as 'up' or 'down'
data['Direction'] = ['Up' if x > 0 else 'Down' for x in data['PriceChange'].values]

# Group volume by direction
volume_by_direction = data.groupby('Direction')['Volume'].sum()

# Plotting the volume profile
plt.figure(figsize=(8, 6))
plt.bar(volume_by_direction.index, volume_by_direction.values, color=['green', 'red'], alpha=0.7)
plt.title('S&P 500 Volume Profile (Bulls vs Bears) - Last 5 Days')
plt.xlabel('Market Direction')
plt.ylabel('Volume')
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Obtendo dados históricos do S&P 500
sp500_data = yf.download('^GSPC', start='2020-01-01', end='2024-01-01')

# Convertendo a coluna de timestamp para datetime
sp500_data.reset_index(inplace=True)
sp500_data['Timestamp'] = pd.to_datetime(sp500_data['Date'])

# Definindo um tipo aleatório para simular as compras e vendas
import random
sp500_data['Type'] = [random.choice(['buy', 'sell']) for _ in range(len(sp500_data))]

# Agrupando os dados por intervalos de 5 minutos e calculando o volume de compra e venda
sp500_data.set_index('Timestamp', inplace=True)
df_5min = sp500_data.resample('5T').sum()

# Filtrando apenas os dados de compra e venda
df_buy = sp500_data[sp500_data['Type'] == 'buy']
df_sell = sp500_data[sp500_data['Type'] == 'sell']

# Agrupando os dados de compra e venda por intervalos de 5 minutos
df_buy_5min = df_buy.resample('5T').sum()
df_sell_5min = df_sell.resample('5T').sum()

# Plotando o gráfico
plt.figure(figsize=(12, 6))
plt.plot(df_5min['Close'], df_5min['Volume'], label='Total Volume', color='blue')
plt.plot(df_buy_5min['Close'], df_buy_5min['Volume'], label='Buy Volume', color='green')
plt.plot(df_sell_5min['Close'], df_sell_5min['Volume'], label='Sell Volume', color='red')

plt.title('Volume de Compra e Venda em Função do Preço (Velas de 5 Minutos)')
plt.xlabel('Preço')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Step 1: Get Historical Stock Data
stock_data = yf.download('AAPL', start='2023-01-01', end='2024-01-01')

# Step 2: Calculate Volume by Price
price_bins = 20  # Number of price ranges
price_range = (stock_data['Close'].max() - stock_data['Close'].min()) / price_bins
volume_by_price = []
for i in range(price_bins):
    lower_bound = stock_data['Close'].min() + i * price_range
    upper_bound = lower_bound + price_range
    volume = stock_data[(stock_data['Close'] >= lower_bound) & (stock_data['Close'] < upper_bound)]['Volume'].sum()
    volume_by_price.append(volume)

# Step 3: Plot the VBP Chart
plt.bar(range(price_bins), volume_by_price)
plt.xlabel('Price Ranges')
plt.ylabel('Volume')
plt.title('Volume by Price (AAPL)')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def volume_profile(data):
    # Define the price levels
    price_levels = data['Close'].unique()

    # Initialize an empty dictionary to store volume at each price level
    volume_profile = {price: 0 for price in price_levels}

    # Aggregate volume at each price level
    for price in price_levels:
        volume_profile[price] = data[data['Close'] == price]['Volume'].sum()

    return volume_profile

# Define the ticker symbol and get intraday data
tickerSymbol = 'AAPL'
tickerData = yf.Ticker(tickerSymbol)
intraday_data = tickerData.history(period='1d', interval='5m')

# Calculate volume profile for 5-minute intervals
volume_profile_5min = volume_profile(intraday_data)

# Convert the volume profile to a DataFrame for easier manipulation
volume_profile_df = pd.DataFrame(volume_profile_5min.items(), columns=['Price', 'Volume'])

# Calculate total volume
total_volume = volume_profile_df['Volume'].sum()

# Calculate percentage distribution of volume relative to price
volume_profile_df['Volume_Percentage'] = volume_profile_df['Volume'] / total_volume * 100

# Plot the percentage distribution of volume as a vertical histogram with annotations
plt.figure(figsize=(8, 10))  # Adjust the figure size as needed
bars = plt.barh(volume_profile_df['Price'], volume_profile_df['Volume_Percentage'], height=0.05)  # Use barh for horizontal bars
plt.xlabel('Percentage of Volume')
plt.ylabel('Price')
plt.title('Percentage Distribution of Volume for 5-Minute Period')
plt.grid(True)

# Add percentage values next to each bar
for bar, percentage in zip(bars, volume_profile_df['Volume_Percentage']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{percentage:.2f}%', ha='left', va='center')

plt.show()
```
