```python
# Função para obter os dados reais do Yahoo Finance
def get_real_data(symbol, start_date, end_date):
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    return stock_data.reset_index()
```

```python
# Defina o caminho para o diretório onde você armazena os arquivos CSV
dir_path = '/content/drive/MyDrive/AI Financial Analisys/Predictions/'

# Lista os arquivos CSV no diretório
csv_files = [f for f in os.listdir(dir_path) if f.endswith('.csv')]

# Loop sobre cada arquivo CSV
for csv_filename in csv_files:
    # Extrai o símbolo do estoque do nome do arquivo CSV
    stock_signal = csv_filename.split('_')[1]

    # Construa o caminho completo para o arquivo CSV
    csv_path = os.path.join(dir_path, csv_filename)

    # Leia o arquivo CSV com as previsões
    predictions_df = pd.read_csv(csv_path)

    # Obtenha os dados reais do Yahoo Finance
    real_data_df = get_real_data(stock_signal, predictions_df['Date'].min(), predictions_df['Date'].max())

    # Crie um DataFrame combinando previsões e dados reais
    combined_df = pd.merge(predictions_df, real_data_df, on='Date', how='inner')
    combined_df.head()

    # Plotar previsões e dados reais
    plt.figure(figsize=(12, 6))
    plt.plot(combined_df['Date'], combined_df['Prediction'], label='Previsão', linestyle='-', marker='o')
    plt.plot(combined_df['Date'], combined_df['Close'], label='Real', linestyle='-', marker='o')

    plt.title(f'Comparação entre Previsão e Real para {csv_filename}')
    plt.xlabel('Data')
    plt.ylabel('Valor')
    plt.legend()
    plt.xticks(rotation='vertical')

    # Adicione isso para mostrar o gráfico
    plt.show()
```
